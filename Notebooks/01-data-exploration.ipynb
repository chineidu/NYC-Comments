````python
# Remove the "By" from the byline
articles_df = articles_df.with_columns(
    byline=pl.col("byline").str.replace_all("By ", "").str.to_lowercase()
)
articles_df["byline"].n_unique()
````

````python
authors_df: pl.DataFrame = (
    articles_df.group_by("byline")
    .agg(
        pl.col("articleWordCount").mean().round(2).alias("avg_word_count"),
        pl.col("articleWordCount").len().alias("num_articles"),
        pl.col("documentType")
        .map_elements(lambda x: list(set(x)))
        .alias("documentType"),
        pl.col("newDesk").map_elements(lambda x: list(set(x))).alias("newDesk"),
        pl.col("sectionName").map_elements(lambda x: list(set(x))).alias("sectionName"),
        pl.col("typeOfMaterial")
        .map_elements(lambda x: list(set(x)))
        .alias("typeOfMaterial"),
    )
    .sort("num_articles", descending=True)
)

authors_df.head()
````

````python
others: list[str] = ["Unknown", "Obits", "SpecialSections", "Summary", "Video"]
opinion_and_editorial: list[str] = [
    "OpEd",
    "Editorial",
    "Upshot",
    "BookReview",
    "Letters",
]
news: list[str] = [
    "National",
    "Metro",
    "Foreign",
    "Washington",
    "Science",
    "Metropolitan",
    "NewsDesk",
    "Climate",
    "Investigative",
    "Politics",
    "NYTNow",
]
business_and_finance: list[str] = [
    "Business",
    "RealEstate",
    "Insider",
    "SundayBusiness",
    "Automobiles",
]
lifestyle_and_culture: list[str] = [
    "Learning",
    "Culture",
    "Magazine",
    "Dining",
    "Games",
    "Well",
    "Sports",
    "Weekend",
    "Travel",
    "Arts&Leisure",
    "Styles",
    "Podcasts",
    "Photo",
    "Smarter Living",
    "Express",
    "TStyle",
    "EdLife",
    "Society",
]


def map_newdesk(newdesk: str) -> str:
    if newdesk in opinion_and_editorial:
        return "Opinion and editorial"
    elif newdesk in news:
        return "News"
    elif newdesk in business_and_finance:
        return "Business and finance"
    elif newdesk in lifestyle_and_culture:
        return "Lifestyle and culture"
    else:
        return "Other"
````

````python
authors_df = authors_df.with_columns(
    newDesk=pl.col("newDesk").map_elements(
        lambda x: list(set([map_newdesk(val) for val in x]))
    )
)

authors_df.head()
````

````python
print(authors_df["documentType"].value_counts(sort=True))
print(authors_df["newDesk"].value_counts(sort=True))
print(authors_df["typeOfMaterial"].value_counts(sort=True))
````

### Comment

- The `newDesk` is dominated by `news`, `opinion and editorial`, `lifestyle and culture` and `business and finance`.
- Most authors write `articles`. Very few write `blogposts` and when theey do, they write both.
- The `sectionName` is dominated by `politics`, `sunday review`, `television` and `asia pacific`.
- On average, the `word_count` is 1286.64 and the average number of articles per author is 3.
- The common type of material is `News`, `OpinionAndEditorial`, `briefing`, `review` and `blog`.

````python
articles_df["typeOfMaterial"].value_counts(sort=True)
````

````python
def map_typeOfMaterial(type_: str) -> str:
    type_ = type_.lower()
    if type_ not in ["news", "op-ed", "review", "editorial", "briefing", "blog"]:
        return "other"
    else:
        return type_
````

````python
authors_df = authors_df.with_columns(
    typeOfMaterial=pl.col("typeOfMaterial").map_elements(
        lambda x: list(set([map_typeOfMaterial(val) for val in x]))
    )
)

authors_df.sample(n=5, seed=5)
````

````python
articles_df = articles_df.with_columns(
    typeOfMaterial=pl.col("typeOfMaterial").map_elements(map_typeOfMaterial)
)

articles_df["typeOfMaterial"].value_counts(sort=True)
````

### Create Article Metadata

- `byline`

````python
articles_df.head(1)
````

````python
# Create `metadata` using: `byline`, `headline`, `keywords`, `newDesk`, `snippet`, `sectionName`
articles_cleaned_df: pl.DataFrame = articles_df.with_columns(
    metadata=pl.concat_str(
        pl.col("byline"),
        pl.col("headline"),
        pl.col("keywords"),
        pl.col("newDesk"),
        pl.col("snippet"),
        separator=" | ",
    )
    .str.to_lowercase()
    .str.replace_all(r"\W+", " "),
    metadata_1=pl.concat_str(
        pl.col("byline"),
        pl.col("headline"),
        pl.col("keywords"),
        pl.col("newDesk"),
        pl.col("snippet"),
        pl.col("sectionName"),
        separator=" | ",
    )
    .str.to_lowercase()
    .str.replace_all(r"\W+", " "),
    metadata_2=pl.concat_str(
        pl.col("byline"),
        pl.col("headline"),
        pl.col("keywords"),
        pl.col("newDesk"),
        pl.col("snippet"),
        pl.col("sectionName"),
        pl.col("documentType"),
        pl.col("source"),
        pl.col("typeOfMaterial"),
        separator=" | ",
    )
    .str.to_lowercase()
    .str.replace_all(r"\W+", " "),
).drop(
    [
        "byline",
        "headline",
        "keywords",
        "newDesk",
        "snippet",
        "sectionName",
        "multimedia",
        "printPage",
        "webURL",
    ]
)
articles_cleaned_df.head()
````
