# Hour (0-23)

````python
# Remove the "By" from the byline
articles_df = articles_df.with_columns(
    byline=pl.col("byline").str.replace_all("By ", "").str.to_lowercase()
)
articles_df["byline"].n_unique()
````

````python
authors_df: pl.DataFrame = (
    articles_df.group_by("byline")
    .agg(
        pl.col("articleWordCount").mean().round(2).alias("avg_word_count"),
        pl.col("articleWordCount").len().alias("num_articles"),
        pl.col("documentType")
        .map_elements(lambda x: list(set(x)))
        .alias("documentType"),
        pl.col("newDesk").map_elements(lambda x: list(set(x))).alias("newDesk"),
        pl.col("sectionName").map_elements(lambda x: list(set(x))).alias("sectionName"),
        pl.col("typeOfMaterial")
        .map_elements(lambda x: list(set(x)))
        .alias("typeOfMaterial"),
    )
    .sort("num_articles", descending=True)
)

authors_df.head()
````
